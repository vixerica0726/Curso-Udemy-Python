### Definindo o número de classes, quando ela não é dada.

#### Regra de Sturges

# $$k = 1 + \frac {10}{3}\log_{10}n$$

```python
n = df_clientes.shape[0]
n
```

```python
# regra de Sturges é um método para definição do número de classes, baseado no total de amostras do banco de dados;
# Calcular o número ideal de classes usando a fórmula de Sturges
k = 1 + (10/3) * np.log10(n)
k = int(k.round(0))
print(f'O número de classes ideal é {k:.2f}')
```

```python
#O parâmetro bins, da função cut(), pode receber o total de classes que desejamos em nossa distribuição de frequências. Neste caso, todas as classes terão a mesma amplitude.
# Criar a tabela de frequências
frequencia = pd.Series(
    pd.cut(x=df_clientes.frequencia_uso_mensal, bins= k, include_lowest=True)
).value_counts()

percentual = pd.Series(
    pd.cut(x=df_clientes.frequencia_uso_mensal, bins= k , include_lowest=True)
).value_counts(normalize=True,  sort = False,)*100

# Exibir as tabelas de frequências
print("Tabela de Frequências:")
print(frequencia)
print("\nTabela de Frequências Normalizada:")
print(percentual)

dist_freq_quantitativas_personalizadas = pd.DataFrame(
    {'Frequência': frequencia, 'Porcentagem (%)': percentual}
)
dist_freq_quantitativas_personalizadas.sort_index(ascending=False)
```

# <font color="#6a0dad" size="10">**5 ANALISE DA AMOSTRA**

Supondo que os dados recebidos são uma amostra da população total de clientes (100000), vamos avaliar se a amostra é adequada para analise.

```python
import math

# Função para calcular a margem de erro
def calcular_margem_erro(n, N, z=1.96, p=0.5):
    """
    Calcula a margem de erro para uma amostra dada.

    Args:
    n (int): Tamanho da amostra.
    N (int): Tamanho da população.
    z (float): Valor z para o nível de confiança (1.96 para 95% de confiança).
    p (float): Proporção esperada (0.5 para máxima variabilidade).

    Returns:
    float: Margem de erro.
    """
    # Calcular a margem de erro usando a fórmula estatística
    margem_erro = z * math.sqrt((p * (1 - p)) / n) * math.sqrt((N - n) / (N - 1))
    return margem_erro

# Função para calcular o tamanho da amostra necessário
def calcular_tamanho_amostra(populacao, confianca, margem_erro):
    """
    Calcula o tamanho da amostra necessário para uma população finita.

    Args:
    populacao (int): Tamanho da população.
    confianca (int): Nível de confiança (90, 95, 99).
    margem_erro (float): Margem de erro desejada (%).

    Returns:
    int: Tamanho da amostra ajustado.
    """
    # Valor z correspondente ao nível de confiança
    z = {90: 1.645, 95: 1.96, 99: 2.576}[confianca]
    # Proporção esperada (0.5 para máxima variabilidade)
    p = 0.5
    # Margem de erro em decimal
    e = margem_erro / 100
    # Calcular o tamanho da amostra inicial usando a fórmula estatística
    n = (z**2 * p * (1 - p)) / e**2
    # Ajustar o tamanho da amostra para a população finita
    n_ajustado = n / (1 + ((n - 1) / populacao))
    return math.ceil(n_ajustado)

# Cenário 1: Calcular o nível de confiança e margem de erro de uma amostra de 200 com população total de clientes (100000)
n = 200
N = 100000
margem_erro = calcular_margem_erro(n, N)
nivel_confianca = 1 - margem_erro

print(f"Cenário 1:")
print(f"Margem de erro para uma amostra de {n} com uma população de {N}: {margem_erro:.2%}")
print(f"Nível de confiança correspondente: {nivel_confianca:.2%}")

# Cenário 2: Calcular o tamanho da amostra necessário para uma população de 100000 clientes com 95% de confiança e 2% de margem de erro
populacao = 100000
confianca = 95
margem_erro_desejada = 2
tamanho_amostra = calcular_tamanho_amostra(populacao, confianca, margem_erro_desejada)

print(f"\nCenário 2:")
print(f"Tamanho da amostra necessário para uma população de {populacao} clientes com {confianca}% de confiança e {margem_erro_desejada}% de margem de erro: {tamanho_amostra}")

# Cenário 3: Calcular o tamanho da amostra necessário para uma população de 100000 clientes com 95% de confiança e 10% de margem de erro
margem_erro_desejada_10 = 10
tamanho_amostra_10 = calcular_tamanho_amostra(populacao, confianca, margem_erro_desejada_10)

print(f"\nCenário 3:")
print(f"Tamanho da amostra necessário para uma população de {populacao} clientes com {confianca}% de confiança e {margem_erro_desejada_10}% de margem de erro: {tamanho_amostra_10}")
```

5 ) Identificar o metodo de amostragem

```python

# Analisar o método de amostragem baseado nas distribuições e variabilidade dos dados
def identificar_metodo_amostragem(data):
    """
    Identifica o método de amostragem baseado nas distribuições e variabilidade dos dados.

    Args:
    data (DataFrame): Dados da amostra.

    Returns:
    str: Método de amostragem identificado.
    """
    # Verificar se a amostra é aleatória simples
    if all(data[column].value_counts(normalize=True).std() < 0.05 for column in data.columns if data[column].dtype == 'object'):
        return "Amostragem Aleatória Simples"

    # Verificar se a amostra é estratificada
    if any(data[column].value_counts(normalize=True).std() >= 0.05 for column in data.columns if data[column].dtype == 'object'):
        return "Amostragem Estratificada"

    # Verificar se a amostra é sistemática
    if any(data[column].value_counts(normalize=True).std() >= 0.10 for column in data.columns if data[column].dtype == 'object'):
        return "Amostragem Sistemática"

    return "Método de Amostragem Desconhecido"

metodo_amostragem = identificar_metodo_amostragem(df_clientes)
print(f"\nMétodo de Amostragem Identificado: {metodo_amostragem}")
```

```python
import pandas as pd
from scipy import stats

# Filtrar os clientes com tempo de assinatura de até 24 meses e separar em 2 grupos: os que fizeram cancelamento e os que não fizeram cancelamento
grupo_cancelaram = df_clientes[(df_clientes['tempo_assinatura_meses'] <= 24) & (df_clientes['cancelou'] == 1)]
grupo_nao_cancelaram = df_clientes[(df_clientes['tempo_assinatura_meses'] <= 24) & (df_clientes['cancelou'] == 0)]

# Variáveis para calcular o Teste T, incluindo 'regiao'
variaveis = ['idade', 'frequencia_uso_mensal', 'mensalidade', 'cancelou', 'regiao', 'tempo_assinatura_meses']

# Dicionário para armazenar os resultados
resultados = {}

# Calcular o Teste T para cada variável
for var in variaveis:
    if var == 'regiao':
        # Converter variáveis categóricas em variáveis dummy
        grupo_cancelaram_dummies = pd.get_dummies(grupo_cancelaram[var], drop_first=True)
        grupo_nao_cancelaram_dummies = pd.get_dummies(grupo_nao_cancelaram[var], drop_first=True)

        # Calcular o Teste T para cada dummy variável
        for dummy_var in grupo_cancelaram_dummies.columns:
            t_stat, p_value = stats.ttest_ind(grupo_cancelaram_dummies[dummy_var].astype(float), grupo_nao_cancelaram_dummies[dummy_var].astype(float))
            resultados[f'regiao_{dummy_var}'] = {'T-statistic': t_stat, 'P-value': p_value}
    else:
        t_stat, p_value = stats.ttest_ind(grupo_cancelaram[var], grupo_nao_cancelaram[var])
        resultados[var] = {'T-statistic': t_stat, 'P-value': p_value}

# Exibir os resultados
for var, res in resultados.items():
    print(f"Variável: {var}")
    print(f"T-statistic: {res['T-statistic']}, P-value: {res['P-value']}\n")
```

 <font color="#ff6600">  Com exceção da variável cancelou, nenhuma das variáveis analisadas apresenta uma diferença estatisticamente significativa entre os clientes que cancelaram e os que não cancelaram. Isso sugere que outros fatores, não analisados aqui, podem estar influenciando o cancelamento.
